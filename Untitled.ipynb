## Intro

To survive in the challenging environment of a global market, organizations must recognize and analyze customer attitudes. To be competitive, organizations must recognize and forecast customer preferences and behaviors to maximize customer retention before their rivals do so. 

This analysis identifies factors that affect customer churn, the single most valuable of an organization's assets. One year's data from call log files relating to 3150 customers were selected randomly from an Iranian mobile operator call-center database.

## Dataset info

Dataset represents randomly collected data from an Iranian telecom company’sdatabase over a period of 12 months. A total of 3150 rows of data, each representing a customer, bear information for 16 columns. The attributes that are in this datasetare call failures, frequency of SMS, number of complaints,number of distinct calls, subscription length, age group,charge amount, type of service, seconds of use, status, and frequency of use.

```python
"""
Call  Failure                - total number of failed calls
Complains                    - total number of complains to the CS
Subscription  Length         - tenure in months
Charge  Amount               - amount customer paid throught their tenure
Seconds of Use               - total number of seconds customer used
Frequency of use             - total number of calls made
Frequency of SMS             - total number of SMS sent
Distinct Called Numbers      - total number of distinct numbers called or messaged
Age Group                    - age group 1 - under 15, 2 - 15 to 25, 3 - 25 to 30, 4 - 30 to 45, 5 - 45 to 55
Tariff Plan                  int64
Status                       int64
Age                          - age
Customer Value               - revenue customer brought to the company
FN                           - false negative (for modelling)
FP                           - false positive (for modelling)
Churn                        - 0 - did not churn, 1 - churned
"""
```

```python
#drop unuseful columns and add new 
churn = churn.drop(['FN','FP'], axis=1)
```

```python
churn['Percentage failed calls'] = churn['Call  Failure'] / churn['Frequency of use']
churn['Percentage SMS usage'] = churn['Frequency of SMS'] / (churn['Frequency of use'] + churn['Frequency of SMS'])
churn['Avg call duration'] = churn['Frequency of use']  / churn['Seconds of Use']
churn['Percentage of complains vs usage'] = churn['Complains'] / (churn['Frequency of use'] + churn['Frequency of SMS'])
churn['Avg nb of complains per month'] = churn['Complains'] / churn['Subscription  Length']
churn['Value per month'] = churn['Customer Value'] / churn['Subscription  Length']
```

```python
# load packages
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns F and P-value
fvalue, pvalue = stats.f_oneway(churn_new['Percentage failed calls'],
       churn_new['Percentage SMS usage'], churn_new['Avg call duration'],
       churn_new['Percentage of complains vs usage'], churn_new['Avg nb of complains per month'])
print(fvalue, pvalue)
# 17.492810457516338 2.639241146210922e-05

# get ANOVA table as R like output
import statsmodels.api as sm
from statsmodels.formula.api import ols
# reshape the d dataframe suitable for statsmodels package 
churn_new_melt = pd.melt(churn_new.reset_index(), id_vars=['index'], value_vars=['Percentage failed calls',
       'Percentage SMS usage', 'Avg call duration',
       'Percentage of complains vs usage', 'Avg nb of complains per month'])
# replace column names
churn_new_melt.columns = ['index', 'treatments', 'value']
# Ordinary Least Squares (OLS) model
model = ols('value ~ C(treatments)', data=churn_new_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```
